```python
import numpy as np
import matplotlib.pyplot as plt


def calculate_charging_time(battery_capacity, charger_power, initial_soc, target_soc, efficiency):
    soc_difference = target_soc - initial_soc
    energy_needed = battery_capacity * soc_difference / 100
    actual_energy = energy_needed / efficiency
    charging_time = actual_energy / charger_power
    return actual_energy, charging_time


def tapered_soc_curve(initial_soc, target_soc, total_time):
    soc_values = np.linspace(initial_soc, target_soc, 200)
    time_values = []

    for soc in soc_values:
        if soc <= 80:
            # Fast charging region
            time_values.append((soc - initial_soc) / (target_soc - initial_soc) * total_time * 0.7)
        else:
            # Slower charging region
            extra = (soc - 80) / (target_soc - 80 + 0.0001)
            time_values.append(total_time * (0.7 + 0.3 * extra))

    return soc_values, time_values


def plot_curve(soc_values, time_values):
    plt.figure()
    plt.plot(time_values, soc_values)
    plt.xlabel("Time (hours)")
    plt.ylabel("State of Charge (%)")
    plt.title("Non-Linear Charging Curve (Tapered Charging)")
    plt.grid(True)
    plt.show()


print("\n====== EV Charging Time Estimator (Advanced) ======")

battery_capacity = float(input("Enter Battery Capacity (kWh): "))
charger_power = float(input("Enter Charger Power (kW): "))
initial_soc = float(input("Enter Initial SOC (%): "))
target_soc = float(input("Enter Target SOC (%): "))
efficiency_percent = float(input("Enter Charging Efficiency (% e.g., 90): "))

efficiency = efficiency_percent / 100

energy, total_time = calculate_charging_time(
    battery_capacity,
    charger_power,
    initial_soc,
    target_soc,
    efficiency
)

hours = int(total_time)
minutes = int((total_time - hours) * 60)

print("\n=========== RESULT ===========")
print(f"Energy Required (with losses): {energy:.2f} kWh")
print(f"Estimated Charging Time: {hours} hours {minutes} minutes")
print("================================")

soc_values, time_values = tapered_soc_curve(initial_soc, target_soc, total_time)

plot_curve(soc_values, time_values)
```
